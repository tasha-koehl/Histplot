# histplot

```python
# displot was old way of creating histograms, now has switched to histplot!
# (Will get an error when you try to use displot)
```

```python
import seaborn as sns
```

```python
# data from seaborn library  
# drop nulls to fix graphs later on
penguins = sns.load_dataset('penguins')
penguins.dropna(inplace=True)
penguins.head()
```

```python
# regular old histogram, 1 variable, being counted
# form - sns.hisplot(df.col)
sns.histplot(penguins.bill_length_mm)
```

```python
# to flop x / y axises, need to specify the y axis
# form - sns.hisplot(y = 'col', data = df)
sns.histplot(y = 'bill_length_mm', 
             data = penguins)
```

```python
# what old version of histplot looked like (using displot)
# kde = kernel density estimate plot, helps to visualize the distribution
sns.histplot(x = 'bill_length_mm', 
             data = penguins,
             kde = True
             )
```

```python
# override default number of bins
sns.histplot(x = 'bill_length_mm', 
             data = penguins, 
             bins = 20) 
```

```python
# set your own custom width of the bins
sns.histplot(x = 'bill_length_mm', 
             data = penguins,
             bins = [30, 40, 45, 50, 60]
             )
```

```python
# decide how wide to make the bins, seaborn will decide how many bins you need
sns.histplot(x = 'bill_length_mm', 
            data = penguins, 
             binwidth = 10
            )
```

```python
# same as above, except you decide the range (can exclude negatives / outliers)
sns.histplot(x = 'bill_length_mm',
             data = penguins, 
             binwidth = 10, 
             binrange = (30, 60)
            )
```

```python
# change y axis statistics
# density like a probability density function
sns.histplot(x = 'bill_length_mm', 
             data = penguins,
             stat = 'density'
            )
```

```python
# probabily, odds that a penguin will have a bill a certain size
sns.histplot(x = 'bill_length_mm', 
             data = penguins,
             stat = 'probability'
            )
```

```python
# addition of categorical variables on color
# here is a little hard to see
sns.histplot(x = 'bill_length_mm', 
             data = penguins, 
             hue = 'species');
```

```python
# lots of overlap?, this shows just outline
# still a little hard to see, the colors overlap and create new colors
sns.histplot(x = 'bill_length_mm', 
             data = penguins, 
             hue = 'species',
             element = 'step'
            )
```

```python
# helps distinguish layers
sns.histplot(x = 'bill_length_mm', 
             data = penguins, 
             hue = 'species',
             element = 'poly'
            )
```

```python
# stacks bars, shows what species proportions each bin is
# hard to tell distribution of variables on top though (easy to see gentoo, but not chinstraop)
sns.histplot(x = 'bill_length_mm', 
             data = penguins, 
             hue = 'species', 
             multiple = 'stack'
            )
```

```python
# scales each bucket to range between 0 - 1, shows proportion of each species in each bucket
# example, 50 is about 50% gentoo and 50% chinstrap
sns.histplot(x = 'bill_length_mm', 
             data = penguins, 
             hue = 'species', 
             multiple = 'fill'
            )
```

```python
# 2 var + color, new, can plot 2 var, similar to jointplot, need x and y value
# darker color means higher amount of values there, like a heat map
# can't see range of data any more (after removing y axis), so can add key to help us understand the values
sns.histplot(x = 'bill_length_mm', 
             y = 'bill_depth_mm', 
             data = penguins,
             cbar=True)
```

```python
# can again add a categorical variable
# BUT very hard to see overlaps
sns.histplot(x = 'bill_length_mm', 
             y = 'bill_depth_mm', 
             data = penguins,
             hue = 'species'
            )
```
